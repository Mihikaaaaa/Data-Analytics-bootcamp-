```python
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    'totalbill':[10,20,30,40,50,15,25,35,45,55],
    'tip':[1,2,3,4,5,1.5,2.5,3.5,4.5,5.5],
    'sex':['Male','Female']*5,
    'smoker':['Yes','No']*5,
    'day':['Thur','Fri','Sat','Sun','Thur','Fri','Sat','Sun','Thur','Fri'],
    'time':['Lunch','Dinner']*5,
    'size':[1,2,3,4,1,2,3,4,2,3]
}

df=pd.DataFrame(data)

sns.set(style='whitegrid')

#BARPLOT
sns.barplot(x='day',y='totalbill',data=df)
plt.title("Barplot")
plt.show()
```

```python
#COUNTPLOT
sns.countplot(x='day',data=df)
plt.title("Countplot")
plt.show()
```

```python
#BOXPLOT
sns.boxplot(x='day',y='totalbill',data=df)
plt.title("Boxplot")
plt.show()
```

```python
#VIOLINPLOT
sns.violinplot(x='day',y='totalbill',data=df)
plt.title("Violinplot")
plt.show()
```

```python
#STRIPPLOT
sns.stripplot(x='day',y='totalbill',data=df)
plt.title("Stripplot")
plt.show()
```

```python
#SWARMPLOT
sns.swarmplot(x='day',y='totalbill',data=df)
plt.title("Swarmplot")
plt.show()
```

```python
#POINTPLOT
sns.pointplot(x='day',y='totalbill',data=df)
plt.title("Pointplot")
plt.show()
```

```python
#BOXENPLOT
sns.boxenplot(x='day',y='totalbill',data=df)
plt.title("Boxenplot")
plt.show()
```
